```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imread
import math
from skimage.util import random_noise
from skimage.filters import rank
from scipy import ndimage
```

# Original Image

```python
original_image=imread("circuit_image.png",False,'L')
```

```python
plt.imshow(original_image,plt.cm.gray)
```

```python
processed_image=original_image.copy()
```

# Adding Gaussian Noise

```python
noise_img = random_noise(processed_image, mode='gaussian')
noise_image=noise_img*255
```

```python
plt.imshow(noise_image,plt.cm.gray)
```

# Applying Arithmetic Mean Filter

```python
kernel_arithmetic_mean = 1/9*np.ones((3,3))
img=ndimage.convolve(noise_image, kernel_arithmetic_mean, mode='constant', cval=0.0)
plt.imshow(img,plt.cm.gray)
```
